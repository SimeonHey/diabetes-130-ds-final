## Split into train/test & Scale
* `from sklearn.model_selection import StratifiedShuffleSplit`
* `from sklearn.preprocessing import StandardScaler`

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler

# Split into train and test
X_all, y_all = df.values, labels
feature_names = df.columns

X_all = StandardScaler().fit_transform(X_all)

sss = StratifiedShuffleSplit(n_splits=1, random_state=42)
train_full_idx, test_idx = next(sss.split(X_all, y_all))


X_train_full, y_train_full = X_all[train_full_idx], y_all[train_full_idx]
X_test, y_test = X_all[test_idx], y_all[test_idx]

# Further split train into train and validation
sss = StratifiedShuffleSplit(n_splits=1, random_state=42, test_size=0.2)

train_idx, dev_idx = next(sss.split(X_train_full, y_train_full))
X_train, y_train = X_train_full[train_idx], y_train_full[train_idx]
X_dev, y_dev = X_train_full[dev_idx], y_train_full[dev_idx]

print(f"Shape of train: {X_train.shape}")
print(f"Shape of dev: {X_dev.shape}")
print(f"Shape of test: {X_test.shape}")
```

## Model fitting

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

def eval_model(model, X_train, y_train, X_test, y_test, verbose=True):
    
    labels_unq = ['<30', '>30', 'NO']
    
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred = model.predict(X_test)
    
    print(f"Model {model}")
    print(f"Accuracy score on train data: {accuracy_score(y_train, pred_train)}")
    print(f"Accuracy score: {accuracy_score(y_test, pred)}")
    
    if verbose:
        print(f"Precision score: {precision_score(y_test, pred, average='weighted')}")
        print(f"Recall score: {recall_score(y_test, pred, average='weighted')}")
        print(f"F1 score: {f1_score(y_test, pred, average='weighted')}")

        conf_mx = confusion_matrix(y_test, pred, labels=labels_unq)
        print(f"Confusion matrix (labels = {labels_unq}:\n{conf_mx}")

        plt.imshow(conf_mx, cmap = "jet")
        plt.colorbar()
        plt.show()
    print("---------------------------------------------------------------\n\n")
```

```python
from sklearn.base import BaseEstimator # No **args, **kwargs
from sklearn.base import TransformerMixin # .fit_transform() 'functionality'
from sklearn.kernel_approximation import RBFSampler

class StupidPredictor(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.will_say = None
    def fit(self, X, y):
        self.will_say = y[0]
    def predict(self, X):
        return np.full_like(X[:, 0], self.will_say, dtype='object')
    
class KernelPredictor(BaseEstimator, TransformerMixin):
    def __init__(self, model):
        self.rbf_features = RBFSampler(gamma=1, random_state=42)
        self.model = model
    def fit(self, X, y):
        return self.model.fit(self.rbf_features.fit_transform(X), y)
    def predict(self, X):
        return self.model.predict(self.rbf_features.fit_transform(X))
    
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier

eval_model(KernelPredictor(SGDClassifier()), X_train, y_train, X_dev, y_dev)
eval_model(StupidPredictor(), X_train, y_train, X_dev, y_dev)
eval_model(LogisticRegression(), X_train, y_train, X_dev, y_dev)
eval_model(SGDClassifier(), X_train, y_train, X_dev, y_dev)
eval_model(GaussianNB(), X_train, y_train, X_dev, y_dev)
eval_model(DecisionTreeClassifier(), X_train, y_train, X_dev, y_dev)
eval_model(RandomForestClassifier(), X_train, y_train, X_dev, y_dev)
eval_model(ExtraTreesClassifier(), X_train, y_train, X_dev, y_dev)
eval_model(GradientBoostingClassifier(), X_train, y_train, X_dev, y_dev)
eval_model(AdaBoostClassifier(), X_train, y_train, X_dev, y_dev)
```

```python
def analyse_cums(df, cols):
    print(len(cols))
    fig, ax = plt.subplots(len(cols), figsize=(7,5*len(cols)))
    num_idx = 0
    
    for col in cols:
        idx_sort = reversed(np.argsort(df[col].value_counts()))
        vals = df[col].value_counts()[idx_sort].values
        cumsum = np.cumsum(vals)
        fullsum = np.sum(vals)
        
        cur_axs = ax[num_idx] if len(cols) > 1 else ax
        
        cur_axs.set_title('Cumulative sum of ' + str(col) + ' values')
        cur_axs.plot(cumsum / fullsum)
        num_idx += 1
    
    return fig

analyse_cums(original, ['diag_1', 'diag_2', 'diag_3']).savefig('Cumulative distributions of diags.png')
analyse_cums(original, ['medical_specialty']).savefig('Cumulative distributions of medspec.png')
```

```python
def log(*args, **kwargs):
    print("Fancy print:", args, kwargs)

def compress_feature(df, col, keep_num):
    compressed_token = 'other'
    idx_sort = reversed(np.argsort(df[col].value_counts()))
    series = df[col].value_counts()[idx_sort]
    
    vals = series.values
    labels = series.index
    
    for lbl in labels[keep_num:]:
        df[col].replace(lbl, compressed_token, inplace=True)
    
play = original.copy()
compress_feature(play, 'diag_1', 100)
compress_feature(play, 'diag_2', 50)
compress_feature(play, 'diag_3', 10)
play
```

```python
def compress_feature_inplace(df, cols, keep_num, isfrac=False):
    if not isfrac:
        log("Keeping top", keep_num, "of columns", cols)
    else:
        log("Keeping top", keep_num, " fraction of columns", cols)
    
    compressed_token = unknown_token
    
    for col in cols:
        idx_sort = reversed(np.argsort(df[col].value_counts()))
        series = df[col].value_counts()[idx_sort]
        
        vals = series.values
        labels = series.index
        
        tarsum = np.sum(vals) * keep_num
        cursum = 0
        
        start_from = 0
        for val in vals:
            if cursum >= tarsum:
                break
            cursum += val
            start_from += 1
        
        log(f"Compressing values from {start_from} to {len(labels)} until the end (cumsum = {cursum})")
        
        for lbl in labels[start_from:]:
            df[col].replace(lbl, compressed_token, inplace=True)
```
